```python
API_KEY = "df57c152a532b88aa2501d0141e9f090"   
BASE_URL = "https://api.openweathermap.org/data/2.5/forecast"
```

```python
cities = [
    # ðŸ‡®ðŸ‡³ India
    "Delhi,IN","Mumbai,IN","Chennai,IN","Kolkata,IN","Bangalore,IN",
    "Hyderabad,IN","Jaipur,IN","Patna,IN","Ranchi,IN","Lucknow,IN",

    # ðŸ‡ºðŸ‡¸ USA
    "New York,US","Los Angeles,US","Chicago,US","Houston,US",
    "Phoenix,US","San Francisco,US","Seattle,US","Miami,US",

    # ðŸ‡ªðŸ‡º Europe
    "London,UK","Paris,FR","Berlin,DE","Rome,IT","Madrid,ES",
    "Amsterdam,NL","Vienna,AT","Zurich,CH",

    # Asia
    "Tokyo,JP","Osaka,JP","Seoul,KR","Beijing,CN","Bangkok,TH",
    "Singapore,SG","Jakarta,ID","Kuala Lumpur,MY",

    # Others
    "Dubai,AE","Doha,QA","Cairo,EG","Cape Town,ZA",
    "Sydney,AU","Melbourne,AU","Toronto,CA","Vancouver,CA"
]
```

```python
def fetch_forecast(city):
    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric"
    }
    
    try:
        response = requests.get(BASE_URL, params=params, timeout=10)
        if response.status_code != 200:
            print("Failed:", city)
            return []
        
        data = response.json()
        records = []

        for item in data["list"]:
            record = {
                "city": city,
                "temperature": item["main"]["temp"],
                "humidity": item["main"]["humidity"],
                "pressure": item["main"]["pressure"],
                "wind_speed": item["wind"]["speed"],
                "weather": item["weather"][0]["main"],
                "timestamp": datetime.fromtimestamp(item["dt"])
            }
            records.append(record)

        print(f"{city} â†’ {len(records)} records collected")
        return records

    except Exception as e:
        print("âš  Error:", city, e)
        return []
```

```python
weather_data = []

for city in cities:
    city_data = fetch_forecast(city)
    weather_data.extend(city_data)
    time.sleep(1)   # small delay for safety
```

```python
df = pd.DataFrame(weather_data)
df.head()
```

```python
df.to_csv("../data/raw_weather_data.csv", index=False)
print("raw_weather_data.csv saved successfully")
df.head()
```
